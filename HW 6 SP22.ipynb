# Decision Trees and Random Forests

## 1. What is a decision tree?

- A decision tree is a type of supervised machine learning

- It is used to categorize or make predictions

- The model is a form of supervised learning, meaning that the model is trained and tested on a set of data that contains the desired outputs. 

- The decision tree may not always provide a clear-cut answer or decision for outputs.

- A decision tree is a map of the possible outcomes of a series of related choices. 

- A decision tree typically starts with a single node, which branches into possible outcomes.

## 2. What are the advantages and disadvantages of a decision tree?

Advantages:

- It can be used for both classification and regression problems: 

- It can capture linear as well as nonlinear relationships

- It requires less data cleaning compare to other techniques

- It is Non-Parametric method so there are no assumptions about the spatial distribution and the classifier structure.

Disadvantages:

- Concerning the decision tree split for numerical variables in numbers of records

- Take more time for training-time complexity to increase as the input increases.

- If the size of data is too big, then one single tree may grow a lot of nodes which might result in complexity and leads to overfitting.

## 3. Explain pruning

- The goal pruning is to optimize the model by eliminating the values of the weight tensors

- The aim is to get a computationally cost-efficient model that takes a less amount of time in training.

- Model with prunning can offer

- It can be used as a regularization technique to prevent overfitting

- A compression mechanism for creating smaller versions of models with marginal depreciation in model performance

- For reducing computational complexity and, in turn, inference time

## 4. What is ensembling?

- Ensemble learning is a approach to machine learning that try to better predictive performance by combining the predictions from multiple models.

- Ensemble methods usually produces more accurate solutions than a single model would.

- The three main classes of ensemble learning methods are bagging, stacking, and boosting

## 5. Explain bagging

- Machine Lachine Learning uses several techniques to build models and improve their performance

- Ensemble learning methods help improve the accuracy of classification and regression models

- The Random Forest algorithm is an example of ensemble learning.

- Bagging, also known as Bootstrap aggregating, is an ensemble learning technique that helps to improve the performance and accuracy of machine learning algorithms

- It is used to deal with bias-variance trade-offs and reduces the variance of a prediction model

- Bagging avoids overfitting of data and is used for both regression and classification models, specifically for decision tree algorithms.

## 6. What are random forests?

- Random Forest is a powerful and versatile supervised machine learning algorithm that grows and combines multiple decision trees to create a forest

- It can be used for both classification and regression problems

- it starts at a single point and then branches off into two or more directions, with each branch of the decision tree offering different possible outcomes

- Random Forest grows multiple decision trees which are merged together for a more accurate prediction.

## 7. Classify the iris dataset using decision trees

```python
# importing librariesf
import pandas as pd
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix 
from sklearn.datasets import load_iris, load_boston
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
iris
df=pd.DataFrame(data=iris['data'],columns=['sepal length (cm)', 'sepal width (cm)','petal length (cm)', 'petal width (cm)'])
df['target']=iris['target']
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:4], df['target'], test_size = 0.25, random_state = 42)
DT_clf = DecisionTreeClassifier(criterion='entropy',max_depth = 3,splitter='best')
DT_clf = DT_clf.fit(X_train,y_train)
```

```python
y_train_preddt = DT_clf.predict(X_train)
print('y_train_pred\n', confusion_matrix(y_train, y_train_preddt))
y_test_preddt = DT_clf.predict(X_test)
print('y_test_pred\n', confusion_matrix(y_test, y_test_preddt))
```

```python
plt.plot(y_train, y_train_preddt, 'o')
plt.xlabel('y_train')
plt.ylabel('y_pred')
```

## 9. Plot the tree and feature importances

```python
tree.plot_tree(DT_clf,feature_names=['sl','sw','pl', 'pw'],class_names=iris['target_names'],filled=True,node_ids=True,proportion=True,rounded=True,fontsize=7)
```

```python
plt.barh(['sl','sw','pl', 'pw'], DT_clf.feature_importances_)
```

## 10. Repeat questions 7 and 8 using random forests

```python
RF_clf = RandomForestClassifier(n_estimators=10,criterion='entropy',max_depth=10,min_samples_split=3,min_samples_leaf=1,max_features='sqrt')
RF_clf = RF_clf.fit(X_train,y_train)
```

```python
y_train_predrf = RF_clf.predict(X_train)
print('y_train_pred\n', confusion_matrix(y_train, y_train_predrf))
y_test_predrf = RF_clf.predict(X_test)
print('y_test_pred\n', confusion_matrix(y_test, y_test_predrf))
```

```python
plt.plot(y_train, y_train_predrf, '*')
plt.xlabel('y_train')
plt.ylabel('y_pred')
```

```python
plt.barh(['sl','sw','pl', 'pw'], RF_clf.feature_importances_)
```
