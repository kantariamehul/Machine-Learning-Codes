# Dimensionality reduction and PCA

## 1. Discuss curse of dimensionality

- The Curse of Dimensionality is too many features
- As the number of features increases the efficiency of the algorithm is generally decreases as the number of dimentions increases which ultimetly increases the possibility of possible answers
- simple examples is if we draw a circle of unit radious the points are in 2D space but if we draw a spear of same radious the space is 3D and there are more space covered
- It is also possible that all features are are not equally important to predicy classes

## 2. Discuss any 3 dimensionality reduction techniques

- The main dimensional reduction techniques are

- Principal component analysis (PCA)
- Linear Discriminatory Analysis (LDA)
- Singular Value Composition
- Isomap embedding
- Locally linear embedding (LLE)


## 3. Explain PCA in your own words

- Principal Component Analysis (PCA) is a statistical technique used for data reduction without losing its properties
- It describes the composition of variances and covariances through several linear combinations of the primary variables
- It is about obtaining a unique set of orthogonal axes where the data has the largest variance
- Its main aim is to overcome the dimensionality of the problem
- The reduction of dimensionality should be such that when dropping higher dimensions, the loss of data is minimum

## 4. Perform classification on wine dataset using any model of your choice using all the features

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
wine = datasets.load_wine()
```

```python
wine
```

```python
X = wine.data
y = wine.target
df_X = pd.DataFrame(X, columns= wine.feature_names)
df_X.head()
```

```python
from sklearn.preprocessing import StandardScaler

df_X= StandardScaler().fit_transform(df_X)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df_X,y,test_size=0.3,random_state=42)
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error

xgb_model = xgb.XGBClassifier(random_state=42)

xgb_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

y_test_pred = xgb_model.predict(x_test)

print(confusion_matrix(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```

## 5. Perform classification on wine dataset using PCA and taking all top components such that 90% variance is explained by them

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 7)
principalComponents = pca.fit_transform(df_X)
Weightages=pca.explained_variance_ratio_
print(Weightages)
Weightages.sum()
```

```python
principalDf = pd.DataFrame(data = principalComponents, columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7'])
principalDf.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(principalDf,y,test_size=0.3,random_state=42)

xgb_model.fit(x_train, y_train)

y_test_pred = xgb_model.predict(x_test)

print(confusion_matrix(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```
