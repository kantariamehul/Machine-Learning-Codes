## 4. Explain gradient descent

*   Gradient descent algorithm is an optimization algorithm
*   It is commonly used to train machine learning models and neural networks
*   It is a method by which cost function can be minimised 
*   Cost function within gradient descent equation has parabola shape
*  With multiple iterations of Cost functions it is possible to find minima point which has minimum cost function value

## 5. What is categorical cross entropy

*   Categorical cross entropy is a loss function that is used in multi-class classification in machine learning 
*  These are tasks where an example can only belong to one out of many possible categories
*  Model must decide which one is the possible class for the given inputs
*  Cross-entropy is a measure from the field of information theory
*  Building upon entropy and generally calculating the difference between two probability distributions.

## 6. Explain in short the following terms: Relu, Sigmoid, softmax

*   Sigmoid function is binary classifier and map outputs between 0 and 1. With help of threshold the output can be converted in binary values.
*   Where as in case of multi class classification sigmoid is not useful hence softmax is used. It gives outputs in terns of probabilities between 0 to 1. The addition of all the outputs is always 1 in case of foftmax.
*   Tanh Function is used to map output between -1 to 1.
*   ReLU (Rectified Linear Units) and Leaky ReLU is being used in most Deep Learning applications like  Computer Vision, Speech Recognition, Natural Language Processing and Deep Neural Networks etc. Softplus (SmoothReLU), Noisy ReLU, Leaky ReLU, Parametric ReLU and ExponentialReLU (ELU) are variations of it.

## 9. pre process the target variable to make it binary

```python
from sklearn.model_selection import train_test_split
Lab=df.iloc[:,0]
print(Lab.head())
y=pd.get_dummies(Lab)
#y.head()
#print(type(y))
#y.value_counts()
X_temp, X_temp1, Lab_train, Lab_test = train_test_split(Lab,Lab,test_size = 0.15,random_state = 42)
```

## 10. construct a neural network

```python

from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils.vis_utils import plot_model

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:],y,test_size = 0.15,random_state = 42)
print(type(y_test))
```

```python
# Build neural network in Keras
epochs_val=5
model = Sequential()

model.add(Dense(784, activation='relu', input_dim = 784))
model.add(Dense(392, activation='relu'))
model.add(Dense(196, activation='relu'))
model.add(Dense(98, activation='relu'))
model.add(Dense(49, activation='relu'))
model.add(Dense(24, activation='relu'))
model.add(Dense(10, activation='sigmoid'))

model.compile(optimizer='rmsprop',loss = 'binary_crossentropy', metrics = ['accuracy'])
history = model.fit(X_train, y_train, epochs = epochs_val)
```
