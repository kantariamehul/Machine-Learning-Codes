## 2. discuss what is word embedding, lemmatization, stemming

Word embedding
- Word Embedding or Word Vector is a numeric vector input that represents a word in a lower-dimensional space
- It allows words with similar meaning to have a similar representation
- They can also approximate meaning
- A word vector with 50 values can represent 50 unique features
- It is used to reduce dimensionality and to use a word to predict the words around it.

Lemmatization
- Lemmatization uses the context in which the word is being used
- lemmatization converts words into their base forms
- In linguistics, lemmatization helps a reader consider a word’s intended meaning instead of its literal meaning
- cats,cat’s,cats’ --> How cat is used

Stemming
- Stemming uses the stem of the word
- Study can be used as studied, studying, will study are forms of stemming
- Stemming algorithms are typically rule-based

## 3. What is TF-IDF?

- In eat food, eat veg, eat non-veg--> Word “eat” is not given that much importance as its frequently available in all the sentences and can have same meaning and so not useful.
- Word2Vec is advance techniquw to convert word in numeric values
- In short TF-IDF is an  algorithm that uses the frequency of words to determine how relevant those words are for NLP

## 5. Compare and Contrast content based vs collaborative recommender systems.

- The content-based approach uses additional information about users and/or items.
- This filtering method uses item features to recommend other items similar to what the
user likes and also based on their previous actions or explicit feedback
- For a movies recommender system, the additional information can be, the age, the sex,
the job or any other personal information for users as well as the category, the main
actors, the duration or other characteristics of the movies are used to determine the use
probable preference.
- The Collaborative filtering method for recommender systems is a method that is solely
based on the past interactions that have been recorded between users and items, in
order to produce new recommendations
- Collaborative Filtering tends to find what similar users would like and the
recommendations to be provided in order to classify the users into clusters of similar
types and recommend each user according to the preference of its cluster
- The main idea that governs the collaborative methods is that through past user-item
interactions when processed through the system, it becomes sufficient to detect similar
users or similar items to make predictions based on these estimated facts and insights.

## 7. What are sparse matrices and how do you create them in python?

```python
# create empty data frame to store all reviews
df=pd.DataFrame(np.zeros([len(negative_fileids)+len(positive_fileids),2]),dtype=int,columns=['review','sentiment'],)
df.head()
```

```python
# add all positive reviews with 1
for i in range(len(positive_fileids)):
  df.iloc[i,0]=movie_reviews.raw(fileids=positive_fileids[i])
  df.iloc[i,1]=1
```

```python
df.info
```

```python
# Cleaning: Convert all upper characters to lower characters
df['review']=df['review'].str.lower()
print(df['review'][0][0:100])
```

```python
# Cleaning: Remove all short forms to full form

Shorts=["i'm","he's","she's","it's","you're","we're","they're","isn't","aren't","wasn't","weren't","don't","doesn't","didn't","i've","you've","we've","they've","haven't","hasn't","hadn't","i'd","he'd","she'd","it'd","you'd","we'd","they'd","can't","couldn't","mustn't","shan't","shouldn't","i'll","he'll","she'll","it'll","you'll","we'll","they'll","won't","wouldn't","let's"]
Fulls=["i am","he is","she is","it is","you are","we are","they are","is not","are not","was not","were not","do not","does not","did not","i have","You have","we have","they have","have not","has not","had not","i had","he had","she had","it had","you had","we had","they had","can not","could not","must not","shall not","should not","i will","he will","she will","it will","you will","we will","they will","will not","would not","let us"]

for i in range(len(Shorts)):
  df['review']=df['review'].replace(Shorts[i],Fulls[i],regex=True)

print(df['review'][0][:100])
```

```python
# Cleaning: Remove all stopwords

stopwrd = stopwords.words('english')
#print(stopwrd)

df['review']= df['review'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwrd)]))
print(df['review'][0][:100])
```

```python
# Cleaning: Remove all punctuation characters

#print('ascii_letters=',string.ascii_letters)
#print('digits=',string.digits)
#print('punctuation=',string.punctuation)
Sets=list(string.punctuation)
#print(Sets)

for c in Sets:
  df['review']=df['review'].str.replace(c,'')
  
print(df['review'][0][:100])
```

```python
# convert all word on steam form
word_stemmer = PorterStemmer()
df['review']= df['review'].apply(lambda x: ' '.join([word_stemmer.stem(word) for word in x.split()]))
print(df['review'][0])
```

# CREATING THE MODEL

```python
from sklearn.feature_extraction.text import CountVectorizer

X = np.array(df['review'].values)
y = np.array(df['sentiment'].values)

cv = CountVectorizer(max_features = 5000)
X = cv.fit_transform(df['review']).toarray()

print("X.shape = ",X.shape)
print("y.shape = ",y.shape)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,y,train_size=0.75,random_state=42)
print('X_train=',X_train.shape)
print('X_test=',X_test.shape)
print('Y_train=',Y_train.shape)
print('Y_test=',Y_test.shape)
```

```python
# Gaussian Naive Bayes (GaussianNB) model
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix

gnb=GaussianNB()
gnb.fit(X_train,Y_train)
Y_train_pred_gnb = gnb.predict(X_train)
print("Gaussian Train accuracy=",accuracy_score(Y_train,Y_train_pred_gnb))
print("Gaussian Train classification_report=\n",classification_report(Y_train,Y_train_pred_gnb))
print("Gaussian Train confusion_matrix=\n",confusion_matrix(Y_train,Y_train_pred_gnb))

Y_test_pred_gnb = gnb.predict(X_test)
print("\nGaussian Test accuracy=",accuracy_score(Y_test,Y_test_pred_gnb))
print("Gaussian Test classification_report=\n",classification_report(Y_test,Y_test_pred_gnb))
print("Gaussian Test classification_report=\n",confusion_matrix(Y_test,Y_test_pred_gnb))
```

```python
#Naive Bayes classifier for multinomial model
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix

mnb=MultinomialNB(alpha=1.0,fit_prior=True)
mnb.fit(X_train,Y_train)
Y_train_pred_mnb = mnb.predict(X_train)
print("Multinomial Train accuracy_score= ",accuracy_score(Y_train,Y_train_pred_mnb))
print("Multinomial Train classification_report=\n",classification_report(Y_train,Y_train_pred_mnb))
print("Multinomial Train confusion_matrix=\n",confusion_matrix(Y_train,Y_train_pred_mnb))

Y_test_pred_mnb = mnb.predict(X_test)
print("Multinomial Test accuracy_score=",accuracy_score(Y_test,Y_test_pred_mnb))
print("Multinomial Test classification_report=\n",classification_report(Y_test,Y_test_pred_mnb))
print("Multinomial Test confusion_matrix=\n",confusion_matrix(Y_test,Y_test_pred_mnb))
```

```python
#Naive Bayes classifier for multivariate Bernoulli model
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix

bnb=BernoulliNB(alpha=1.0,fit_prior=True)
bnb.fit(X_train,Y_train)
Y_train_pred_bnb = bnb.predict(X_train)
print("Bernoulli Train accuracy_score= ",accuracy_score(Y_train,Y_train_pred_bnb))
print("Bernoulli Train classification_report=\n",classification_report(Y_train,Y_train_pred_bnb))
print("Bernoulli Train confusion_matrix=\n",confusion_matrix(Y_train,Y_train_pred_bnb))

Y_test_pred_bnb = mnb.predict(X_test)
print("Bernoulli Test accuracy_score= ",accuracy_score(Y_test,Y_test_pred_bnb))
print("Bernoulli Test classification_report=\n",classification_report(Y_test,Y_test_pred_bnb))
print("Bernoulli Test confusion_matrix=\n",confusion_matrix(Y_test,Y_test_pred_bnb))
```
