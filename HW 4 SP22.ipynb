# Classification Algorithms: Logistic Regression and Support Vector Machine

## 1 Explain what is classification and how it is different from regression

Regression use a virtual line to map input features ti ouutpu lables which are numeric values. For example of hight is 5 feet the what will the weight?

While classification is used to classify all input values in multiple classes of the lables. For example weather the person is a boy or girl, weather the given pictire is of animal, bird, human, etc.

So in simple words
To find quantity we use regression
to find class (group) we use classification

## 2 Explain what is Logistic regression, its working and how it is different from linear regression

Logistic regression is a classification technique, which is used to divide data in various classes. It is mostly used for binary classification. 


Normally in linear regression we use a virtual stright line to map the values, while in logistic regression we use signum or softmax function to map output values in 0 or 1 and probability respectively. 

## 3 Explain what is Linear SVM and its working

Support vector machine is a liner classification and regression technique. It identify the plane or line which can divide data set in to to or multiple parts.

It also find the separator line sucg that the 1st nearest points on eithe side has maximum distance from the line/plane. The points which are closeset to the line are known as support vectors. The distance between two support vectors on eithr side is known as margin.

## 4 What do you mean by kernel functions?

In many clases data can not be separated by a line or a plane. So it is not possible to apply linear classification techniques.

Kernel Functions are the set of the transfer functions which can map input data space to some other dimentional data space where it is easily possible to differentiatle different class by a plane in miltiple domention space.

It is used with SVM (Support Vector Machine) to map the data in separable form.

## 5 Discuss how SVM makes use of kernel functions

Kernel Function is et of functions whch are used to take data as input and transform it into the required form of processing data. It is used due to a set of mathematical functions used in Support Vector Machine providing the method to manipulate the data.

So, Kernel Function generally transforms the training set so that a decision surface is able to transform to a linear equation where dimensions are increased sompare to input data. 

Polynomial kernel, Gaussian kernel, Gaussian radial basis function (RBF), Laplace RBF kernel, Hyperbolic tangent kernel, Sigmoid kernel are some of the examples of various kernel functions.

## 6 Discuss the following terms: Accuracy, Precision, Recall, F1 score, Specificity, Sensitivity, AUROC, PRAUC

Accuracy: It measures how many observations, both positive and negative, were correctly identified.

Precision: Precision is the ratio of the correctly positive labeled by our program to all positive labeled data.

Recall: Precision is the ratio of the correctly positive labeled by diabetic in reality.

F1 score: it combines precision and recall into one metric by calculating the harmonic mean between those two. 

Specificity:Specificity is the correctly negative labeled by the program to all who are healthy in reality.

Sensitivity: Sensitivity is the metric that evaluates a model's ability to predict true positives of each available category.

AUROC: AUC stands for "Area under the ROC Curve. The Area Under the Curve (AUC) is the measure of the ability of a classifier to distinguish between classes and is used as a summary of the ROC curve.

PRAUC: It is a curve that combines precision (PPV) and Recall (TPR) in a single visualization. For every threshold, you calculate PPV and TPR and plot it. The higher on y-axis your curve is the better your model performance. This can be interpreted as the probability that a random positive is assigned a higher score than a random negative.
